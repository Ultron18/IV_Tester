This code is to generate the parameters such as front resistance network, back resistance network, rshunt, rseries etc

```python
import numpy as np
import pickle
```

```python
m = 120   # No of fingers
n = 9   # No of busbars

I = 0.01047087 # 0.009 #illumination

# now we set the range for each parametes
RF_finger = []
RFf = []
RB_finger = []
RBf = []
for i in range(0, m):
    for j in range(0, n+1):
        # RFf.append(np.random.uniform(0.1e-3, 15e-3))
        # RBf.append(np.random.uniform(0.1e-3, 15e-3))
        RFf.append(np.random.uniform(1.376, 2.064))
        RBf.append(np.random.uniform(0.7, 1.05))        
    RF_finger.append(RFf)
    RB_finger.append(RBf)
    RFf = []
    RBf = []

RF_busbar = []
RFb = []
RB_busbar = []
RBb = []
for i in range(0, m+1):
    for j in range(0, n):
        # RFb.append(np.random.uniform(0.01e-3, 1.5e-3))
        # RBb.append(np.random.uniform(0.01e-3, 1.5e-3))
        RFb.append(np.random.uniform(0.06075, 0.108))
        RBb.append(np.random.uniform(0.0055, 0.0077))
    RF_busbar.append(RFb)
    RB_busbar.append(RBb)
    RFb = []
    RBb = []

Rshunt = []
Rseries = []
Rsh = []
Rsr = []
# for i in range(0,n):
#     for j in range(0,m):  
#         Rsh.append(np.random.uniform(10, 20000))
#         Rsr.append(np.random.uniform(5e-5, 150e-2))
#     Rshunt.append(Rsh)
#     Rseries.append(Rsr)
#     Rsh = []
#     Rsr = []

for i in range(0,m):
    for j in range(0,n):  
        Rsh.append(np.random.uniform(9800, 10000))
        Rsr.append(np.random.uniform(4.25e-7, 4.3e-7))
    Rshunt.append(Rsh)
    Rseries.append(Rsr)
    Rsh = []
    Rsr = []


#probing coordinates
# num_probe_points_top = 8
# num_probe_points_bottom = 6
# # Generate probe points
# Probe_pos = []
# Probe_neg = []
# # For each column, generate 8 equally spaced points
# for col in range(1, n + 1):
#     row_indices_top = np.linspace(1, m, num_probe_points_top, dtype=int)
#     row_indices_bottom = np.linspace(1, m, num_probe_points_bottom, dtype=int)

#     for row in row_indices_top:
#         Probe_pos.append((row,col))

#     for row in row_indices_bottom:
#         Probe_neg.append((row,col))


num_probe_points = 8
# Generate probe points
Probe_bottom = []
# For each column, generate 8 equally spaced points
for col in range(1, n + 1):
    row_indices = np.linspace(1, m, num_probe_points, dtype=int)

    for row in row_indices:
        Probe_bottom.append((row,col))

# Probe_bottom = [(60,5)]

print("Probes_bottom:", Probe_bottom)

# Probe_bottom = [(60,5)]


num_probes_per_busbar = 8  # Number of equally spaced probe points per column
# Initialize the lists for storing probe points
V_Probe_top = []  # For 3rd and 6th points (Voltage probes)
F_Probe_top = []  # For the rest (Current probes)

# Generate the probe points for each column
for col in range(1, n + 1):
    # Generate 8 equally spaced row indices for the current column (1 to m)
    row_indices = np.linspace(1, m, num_probes_per_busbar, dtype=int)

    # Loop over the 8 row indices for the current column
    for idx, row in enumerate(row_indices):
        # Check if it's the 3rd or 6th probe point (indexing starts from 0)
        if idx == 2 or idx == 5:  # 3rd and 6th (0-based index: 2 and 5)
            V_Probe_top.append((row, col))  # Append to voltage probe list
        else:
            F_Probe_top.append((row, col))  # Append to current probe list

# Display the result
print("V_Probe_top:", V_Probe_top)
print("I_Probe_top:", F_Probe_top)


# pogo pin resistance
Rpogo = 1e-3 #0.1 m ohm
```

```python
# Store all variables in a dictionary
data = {
    "n": n,
    "m": m,
    "I": I,
    "RF_finger": RF_finger,
    "RB_finger": RB_finger,
    "RF_busbar": RF_busbar,
    "RB_busbar": RB_busbar,
    "Rshunt": Rshunt,
    "Rseries": Rseries,
    "RPogo_pin": Rpogo,
    "Probe_bottom": Probe_bottom,
    "V_Probe_top": V_Probe_top,
    "F_Probe_top": F_Probe_top
}

# Save the dictionary to a file
#file_name = "spice_models\\parameters.pkl"
file_name = "Model\\parameters.pkl"
with open(file_name, "wb") as file:
    pickle.dump(data, file)
```
