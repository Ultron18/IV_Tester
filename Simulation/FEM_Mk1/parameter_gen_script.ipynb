This code is to generate the parameters such as front resistance network, back resistance network, rshunt, rseries etc

```python
import numpy as np
import pickle
```

```python
m = 120   # No of fingers
n = 9   # No of busbars

I = 0.01047087 # 0.009 #illumination

# now we set the range for each parametes
RF_finger = []
RFf = []
RB_finger = []
RBf = []
# for j in range(0, m+1):
#     for i in range(0, n):
#         RFf.append(np.random.uniform(0.1e-3, 15e-3))
#         RBf.append(np.random.uniform(0.1e-3, 15e-3))
#     RF_finger.append(RFf)
#     RB_finger.append(RBf)
#     RFf = []
#     RBf = []

for i in range(0, m):
    for j in range(0, n+1):
        # RFf.append(np.random.uniform(0.1e-3, 15e-3))
        # RBf.append(np.random.uniform(0.1e-3, 15e-3))
        RFf.append(np.random.uniform(1.376, 2.064))
        RBf.append(np.random.uniform(0.7, 1.05))        
    RF_finger.append(RFf)
    RB_finger.append(RBf)
    RFf = []
    RBf = []

RF_busbar = [] #RF_busbar (0.05m, 0.15m) ohm
RFb = []
RB_busbar = []
RBb = []
# for i in range(0, n+1):
#     for j in range(0, m):
#         RFb.append(np.random.uniform(0.01e-3, 1.5e-3))
#         RBb.append(np.random.uniform(0.01e-3, 1.5e-3))
#     RF_busbar.append(RFb)
#     RB_busbar.append(RBb)
#     RFb = []
#     RBb = []

for i in range(0, m+1):
    for j in range(0, n):
        # RFb.append(np.random.uniform(0.01e-3, 1.5e-3))
        # RBb.append(np.random.uniform(0.01e-3, 1.5e-3))
        RFb.append(np.random.uniform(0.06075, 0.108))
        RBb.append(np.random.uniform(0.0055, 0.0077))
    RF_busbar.append(RFb)
    RB_busbar.append(RBb)
    RFb = [] 
    RBb = [] 


Rshunt = []
Rseries = []
Rsh = []
Rsr = []
# for i in range(0,n):
#     for j in range(0,m):  
#         Rsh.append(np.random.uniform(10, 20000))
#         Rsr.append(np.random.uniform(5e-5, 150e-2))
#     Rshunt.append(Rsh)
#     Rseries.append(Rsr)
#     Rsh = []
#     Rsr = []

for i in range(0,m):
    for j in range(0,n):  
        Rsh.append(np.random.uniform(9800, 10000))
        Rsr.append(np.random.uniform(4.25e-7, 4.3e-7))
    Rshunt.append(Rsh)
    Rseries.append(Rsr)
    Rsh = []
    Rsr = []


#probing coordinates.
num_probe_points = 8
# Generate probe points
Probe_pos = []
Probe_neg = []
# For each column, generate 8 equally spaced points
for col in range(1, n + 1):
    row_indices = np.linspace(1, m, num_probe_points, dtype=int)

    for row in row_indices:
        Probe_pos.append((row,col))
        Probe_neg.append((row,col))

# pogo pin resistance
Rpogo = 1e-3 #0.1 m ohm
```

```python
# Store all variables in a dictionary
data = {
    "n": n,
    "m": m,
    "I": I,
    "RF_finger": RF_finger,
    "RB_finger": RB_finger,
    "RF_busbar": RF_busbar,
    "RB_busbar": RB_busbar,
    "Rshunt": Rshunt,
    "Rseries": Rseries,
    "RPogo_pin": Rpogo,
    "Probe_pos": Probe_pos,
    "Probe_neg": Probe_neg
    # "R_chuck"  : R_chuck
}

# Save the dictionary to a file
#file_name = "spice_models\\parameters.pkl"
file_name = "Model\\parameters.pkl"
with open(file_name, "wb") as file:
    pickle.dump(data, file)
```
