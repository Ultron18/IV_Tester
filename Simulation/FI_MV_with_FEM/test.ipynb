```python
import matplotlib.pyplot as plt
import ltspice
```

```python
lt1 = ltspice.Ltspice('IVTesterFullCircuit - FIMV_single -10k.raw')
lt1.parse()

lt11 = ltspice.Ltspice('IVTesterFullCircuit - FIMV_single-1k.raw')
lt11.parse()

lt2 = ltspice.Ltspice('IVTesterFullCircuit - FIMV_double.raw')
lt2.parse()
```

```python
Vmeas1 = lt1.get_data('V(vmeas)')
I1 = lt1.get_data('I(R3)')

Vmeas11 = lt11.get_data('V(vmeas)')
I11 = lt11.get_data('I(R3)')

Vmeas2 = lt2.get_data('V(vmeas)')
I2 = lt2.get_data('I(R3)')
```

```python
plt.plot(Vmeas1,I1,label='Single ended measurement 10k')
# plt.plot(Vmeas11,I11,label='Single ended measurement 1k')
plt.plot(Vmeas2,I2,label='Differential measurement')
plt.title('Voltage measurement method comparison')
# plt.xlim([0,0.8])
plt.legend()
```
